```python
pollution=pd.read_csv('outdoor-pollution-death-rate.csv') 
pollution
```

```python
world_pop=pd.read_csv('world-population-by-world-regions-post-1820.csv') 
world_pop
```

```python
energy_pollution=energy.merge(pollution, how='left', on=['Entity', 'Year'])
energy_pollution.columns
```

```python
energy_pollution=energy_pollution[['Entity', 'Year', 'Oil Consumption - EJ',
       'Gas Consumption - EJ', 'Coal Consumption - EJ',
       'Solar Consumption - EJ', 'Hydro Consumption - EJ',
       'Nuclear Consumption - EJ', 'Wind Consumption - EJ',
       'Geo Biomass Other - EJ', 'Biofuels (TWh)', 'Nuclear_Energy_saved',
       'Death rate â€“ Outdoor air pollution (age-standardized) (IHME)']]
```

```python
energy_max=energy_pollution.merge(world_pop, how='left', on=['Entity', 'Year'])
energy_max=energy_max[['Entity', 'Year', 'Oil Consumption - EJ', 'Gas Consumption - EJ',
       'Coal Consumption - EJ', 'Solar Consumption - EJ',
       'Hydro Consumption - EJ', 'Nuclear Consumption - EJ',
       'Wind Consumption - EJ', 'Geo Biomass Other - EJ', 'Biofuels (TWh)',
       'Nuclear_Energy_saved',
       'Death rate â€“ Outdoor air pollution (age-standardized) (IHME)',
       'Total population (Gapminder, HYDE & UN)']]
```

```python
energy_max['pollution_deaths']=0.85*energy_max['Death rate â€“ Outdoor air pollution (age-standardized) (IHME)']*energy_max['Total population (Gapminder, HYDE & UN)']/100000
```

```python
energy_max['pollution_deaths']=energy_max['pollution_deaths'].fillna(0)
energy_max['pollution_deaths']=energy_max['pollution_deaths'].astype(int)
energy_max
```

```python
energy_max.columns
```

```python
energy_max['Air_Pollution_Death_Rate']=energy_max['pollution_deaths']/(energy_max['Oil Consumption - EJ']+energy_max['Gas Consumption - EJ']+energy_max['Coal Consumption - EJ'])
energy_max.iloc[[1154]]
```

```python
energy_max['Nuclear_Energy_saved']=round(energy['Nuclear Consumption - EJ']*energy_max['Air_Pollution_Death_Rate'], 2)
energy_max.iloc[[1154]]
```

```python
energy_max['Perc_from_Nuclear']=round(100*energy_max['Nuclear Consumption - EJ']/(energy_max['Oil Consumption - EJ']+energy_max['Gas Consumption - EJ']+energy_max['Coal Consumption - EJ']+energy_max['Solar Consumption - EJ']+energy_max['Hydro Consumption - EJ']+energy_max['Wind Consumption - EJ']+energy_max['Geo Biomass Other - EJ']+energy_max['Biofuels (TWh)']), 2)
energy_max.iloc[[1154]]
```

```python
energy_max['Death_Reduction_Ratio']=round(100*energy_max['Nuclear_Energy_saved']/(energy_max['Nuclear_Energy_saved']+energy_max['pollution_deaths']), 2)
energy_max.iloc[[1154]]
```
